```python
from IPython.display import Image, display
from ultralytics import YOLO
import os

RUN_DIR = "runs/detect/train"

print("="*50)
print("FILES IN RUN DIRECTORY")
print("="*50)
!ls {RUN_DIR}

print("\n" + "="*50)
print("METRICS")
print("="*50)

model = YOLO(f"{RUN_DIR}/weights/best.pt")
metrics = model.val(data="/content/data.yaml", verbose=False)

print(f"mAP50      : {metrics.box.map50:.4f}")
print(f"mAP50-95   : {metrics.box.map:.4f}")
print(f"Precision  : {metrics.box.mp:.4f}")
print(f"Recall     : {metrics.box.mr:.4f}")

print("\n" + "="*50)
print("PLOTS")
print("="*50)

plots = [
    "results.png",
    "confusion_matrix.png",
    "confusion_matrix_normalized.png",
    "PR_curve.png",
    "F1_curve.png",
    "P_curve.png",
    "R_curve.png"
]

for plot in plots:
    path = f"{RUN_DIR}/{plot}"
    if os.path.exists(path):
        print(f"\n{plot}")
        display(Image(path))
    else:
        print(f"{plot} not found")
```
